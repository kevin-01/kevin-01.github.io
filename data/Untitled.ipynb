```python
with open("combined_data_1.txt") as f:
    content = f.readlines()
content = [x.strip() for x in content]
content[0]
```

```python
import pandas as pd
df = pd.read_csv('movie_titles.csv').set_index("Number")
df
```

```python
res = []
len_ = len(content)
print(len_)
acc = 0

for i in range(len_):
    splitted = content[i].split(",")
    if(i % 100000 == 0):
        print(i)
    if len(splitted) == 1:
        temp = content[i].split(":")
        acc = 0
    elif(acc < 100):
        acc += 1
        idx_movie = int(temp[0])
        name = df.iloc[idx_movie].Name
        res.append(splitted[0] +"," + name + "," + splitted[1])
```

```python
res
```

```python
user = []
movie = []
rating = []
for row in res:
    splitted = row.split(",")
    user.append(splitted[0])
    movie.append(splitted[1])
    rating.append(splitted[2])

print(len(list(set(user))))
print(len(list(set(movie))))
```

```python
s = pd.DataFrame(index=list(set(user)), columns=list(set(movie)))
```

```python
for i in range(len(user)):
    row_ = user[i]
    col_ = movie[i]
    value_ = rating[i]
    s.loc[row_][col_] = value_
s = s.fillna(0)
s.head()
```

```python
s2 = s.head(20)
s2 = s2.loc[:, (s2 != 0).any(axis=0)]
print(len(s2.columns.values))
s2.to_csv('out.csv')
```
